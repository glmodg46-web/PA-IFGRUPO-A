```python
#Exercicio 9
#Entrada
ano = int(input("Digite um ano: "))

#processamento / saida
# 1. divisivel por 400 -> Bissexto
# 2. divisivel por 4 e não por 100 -> Bissexto
# 3. caso contrario -> Não é bissexto
if (ano % 400 == 0) or (ano % 4 == 0 and ano % 100 != 0):
    print(f"{ano} é um ano bissexto.")
else:
    print(f"{ano} não é um ano bissexto.")
```
