Calcular A àrea do retangulo

```python
#Entrada
comprimento =  float(input("Digite o Comprimento do retángulo:"))
altura = float(input("Digite a altura do retángulo:"))


#Processamento

Area = comprimento * altura



#saida

print(f"O total da área do retángulo é:{Area}")
```

Fazer um Programa que calcula o dobro de um número

Entrada:Um numero

Processamento:Calcular o Dobro

Saida:Mostrar o resultado

```python
# Entrada

Numero = float(input("Coloque o número que você quer o dobro: "))

# Processamento

Numero = Numero * 2

#Saída

print(f"o total da multiplicação é: {Numero}")
```

Escrever um programa que calcula a média dos alunos
Entrada:Nome, idade e as 4 notas do bimestre
Processamento:Calcular a média
saída:Mostrar o nome dos alunos e a média obtida


```python
#Entrada

Nome = str(input("Escreva Seu nome?"))

idade = int(input("Qual a sua idade?"))

Responda = str(input("Qual a média que você obteve?"))

primeiro = float(input("Qual a sua primeira nota?"))

segundo = float(input("Qual a sua segundo nota?"))

terceiro = float(input("Qual a sua terceiro nota?"))

quarto = float(input("Qual a sua quarto nota?"))

#Processamento

Total = (primeiro + segundo + terceiro + quarto) / 4

#saída

print(f"Olá {Nome} A média de todos os bimestres é : {Total}")
```
