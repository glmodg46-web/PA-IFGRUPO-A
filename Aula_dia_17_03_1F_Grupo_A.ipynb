```python
# Operadores aritiméticos
# Fazer um programa que calcula operações com 02 numeros


# Entrada

a = int (input("digite um numero: "))
b = int (input("digite um numero: "))


# Processamento

Subtracao  = a - b

Divisao  = a / b

Multiplicação = a * b

Soma = a + b

Resto = a % b

Potencia = a ** b

# Saida
print ("")
print("------------------------OPERADORES ARITIMÉTICOS-----------------------")
print("")
print("----------------------------Subtracao--------------------------------------")
print("")
print (f"O resultado da Subtracaoo de {a} - {b} é igual a : {Subtracao}")
print("")
print("----------------------------------------------------------------------")
print ("")
print("")
print("----------------------------Divisao--------------------------------------")
print("")
print (f"O resultado da Divisao de {a} / {b} é igual a : {Divisao}")
print("")
print("----------------------------------------------------------------------")
print ("")
print("")
print("----------------------------Multiplicação--------------------------------------")
print("")
print (f"O resultado da Multiplicação de {a} * {b} é igual a : {Multiplicação}")
print("")
print("----------------------------------------------------------------------")
print ("")
print("")
print("----------------------------Resto--------------------------------------")
print("")
print (f"O resultado da Divisao de {a} Dividido {b} é igual a : {Resto}")
print("")
print("----------------------------------------------------------------------")
print ("")
print("")
print("----------------------------potenciação--------------------------------------")
print("")
print (f"O resultado da Potenciaçao de {a} ** {b} é igual a : {Potencia}")
print("")
print("----------------------------------------------------------------------")
```
