```python
# Escrever um programa que verifica se a palavra é palindroma e imprimi a palavra

def verificar_palindromo(palavra):
    # Entrada
    palavra_limpa = palavra.replace(" ", "").lower()

    # Processamento
    palavra_invertida = palavra_limpa[::-1]

    # Saida
    if palavra_limpa == palavra_invertida:
        print(f"A palavra '{palavra}' é um palíndromo.")
    else:
        print(f"A palavra '{palavra}' não é um palíndromo.")

entrada = input("Digite uma palavra: ")
verificar_palindromo(entrada)

```

```python
# Escrever um programa que o usuario digita um numero e o programa imprime os numeros pares de zero até o numero

#Entrada

numero = int(input("Digite um número: "))

for i in range(0, numero + 1):
    if i % 2 == 0:
        print(i)
```
